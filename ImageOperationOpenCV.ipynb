# Resizing image

# Augmentation

```python
from imgaug import augmenters as iaa
import matplotlib.pyplot as plt
```

```python
plt.imshow(img)
plt.show()
```

```python
seq=iaa.Sequential([
      iaa.Fliplr(0.5),
      iaa.Flipud(0.5),
      iaa.Affine(rotate=(-45,45))
])
```

```python
augmented_images=seq.augment_image(img)
plt.imshow(augmented_images)
plt.show()
```

```python
plt.figure(figsize=(10,10))
plt.subplot(1,2,1)
plt.imshow(img)
plt.subplot(1,2,2)
plt.imshow(augmented_images)
plt.show()
```

grayscal convert

```python
grayscal_img=cv2.cvtColor(img,cv2.COLOR_BGR2GRAY)
plt.imshow(grayscal_img,cmap='gray')
plt.show()
```

#Normalization

```python
normalized_img=cv2.normalize(img,None,0,255,cv2.NORM_MINMAX) # normalized between 0-255
plt.imshow(normalized_img)
plt.show()
```
